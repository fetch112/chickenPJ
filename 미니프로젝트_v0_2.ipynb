```python
import pandas as pd
import numpy as np

class Read_Write:
    excel_exts = ["xlsx","xlsm","xlsb","xltx","xltxm"]
    
    def __init__(self, path):
        self.path = path
        self.dirname, self.filename  = path.rsplit('/', maxsplit=1)
        self.ext = self.filename.split(".")[1]
        print(self.path, self.filename, self.ext, sep='\n')
        self.df = self.getDF() #초기화 하면서 자신의 메서드 실행 get df 
    
    def getDF(self): 
        if self.ext=="csv": #확장자가 csv다?
            return pd.read_csv(self.path,encoding='CP949')
        elif self.ext in Read_Write.excel_exts: # 확장자가 엑셀이다?
            return pd.read_excel(self.path)
        


################ ###################write####################################
#csv로 저장
    def csv(self, name, *, path=None): # name, path ="/content/drive/My Drive/chickenPJ/data"
        fname = f'{path if path else self.dirname}/{name}.csv'
        print(fname)
        self.df.to_csv(fname)
##엑셀로 저장        
    def excel(self, name, *, path=None): #키워드 온리 뒤에 path 는 키워드 온리
        fname = f'{path if path else self.dirname}/{name}.xlsx'
        print(fname)
        self.df.to_excel(fname)


path="/content/drive/My Drive/chickenPJ/data" #경로

#####1)모든 서울시의 2010년부터 현재까지  영/폐업 
tot = Read_Write(path+"/서울시 법정구역 읍면동 공간정보 (좌표계_ WGS1984).csv")  

tot.df #tot란 인스턴스의 데이터 프레임

```

```python
import pandas as pd
import numpy as np

####################### 정제 클래스##########################
#df_tot      -1)  2010~2020 영폐업치킨
#df_pre_ch   -2)  2020 영업치킨
#df_pop      -3)인구
#df_code     -4) 법정동코드
#df_xy       -5) 구별 동별 좌표

class Refine:
  @staticmethod
  def tot(df,df_code): #1)df_tot 정제
    #-0 df_tot : 영업/폐업 df의 쓸모없는 컬럼 빼기
    df=df[
          ['인허가일자', '영업상태명', '상세영업상태명', '폐업일자',
          '소재지면적','소재지전체주소','도로명전체주소', '사업장명','업태구분명',
          '좌표정보(X)','좌표정보(Y)','시설총규모']
          ]
    #-1 컬럼명 바꾸기
    df.rename({"사업장명":"상호명"},inplace=True)
    #-2 창폐업일자 nan 값 대치 #-3 폐업일자타입변경
    df["인허가일자"]=df["인허가일자"].fillna(0).astype(np.int64)
    df["폐업일자"]=df["폐업일자"].fillna(0).astype(np.int64)
    #-3 설립연도,폐업년도 컬럼 추가
    df["설립연도"]=df["인허가일자"].apply(lambda x: x//10000)
    df["폐업연도"]=df["폐업일자"].apply(lambda x: x//10000)
    #-4 시군구 컬럼 추가
    df["시군구명"]=df["소재지전체주소"].str.split().str[1]
    df["동리명"]=df["소재지전체주소"].str.split().str[2]
    #-5 법정동명과 맞는 행정동 코드 뽑기!
    key1=df[["동리명","소재지전체주소"]] #df_code와 조인해서 행정동코드 뽑아내기
    key1=key1.join(df_code.set_index("동리명")["행정동코드"],on="동리명").drop_duplicates("소재지전체주소")
    df["행정동코드"]=key1["행정동코드"]
    df=df.dropna()
    df["행정동코드"]=df["행정동코드"].astype(np.int64)
    return df


  @staticmethod
  def pre_ch(df):#2)df_pre_ch 정제
    #)df_pre_ch : 현재 영업중인 (2020년기준) 치킨집  => 쓸모없는 컬럼 빼기
    df=df[
          ['상호명','시군구명','법정동명','행정동코드','지번주소','도로명주소','경도','위도']
          ]
    df["행정동코드"]=(df["행정동코드"]/100).astype(np.int64)
    df=df.rename({"법정동명":"동리명"},axis=1)
    return df


  @staticmethod
  def pop(df): #3) df_pop 정제
    #)서울시 인구 구별 데이터 정제 (컬럼명도 변경)
    df=df.drop("행정구역(시군구)별(1)",axis=1)
    df=df.drop([0,1],axis=0)
    df.rename({"행정구역(시군구)별(2)":"시군구명"},axis=1,inplace=True)
    return df

  ##좌표계 메서드
  @staticmethod
  def code(df): #4) df_code 정제
    df=df[df_code.df["시도명"]=="서울특별시"]
    df["행정동코드"]=(df["행정동코드"]/100).astype(np.int64)
    df=df[["행정동코드","동리명"]]
    return df

  @staticmethod
  def xy(df): #1)df_xy 정제 구,동별 좌표계
    df=df.rename({"시군구명_한글":"시군구명"},axis=1)
    df=df[["시군구명","위도","경도"]]
    return df

  @staticmethod
  def dong_xy(df): #1)df_xy 정제 구,동별 좌표계
    df=df[["읍면동명","위도","경도"]]
    df_code=Read_Write("/content/drive/My Drive/chickenPJ/data/행정동 및 법정동 코드.xlsx")
    df_code.df=df_code.df[df_code.df["시도명"]=="서울특별시"]
    df=df.rename({"읍면동명":"동리명"},axis=1)
    df_code.df=df_code.df.rename({"읍면동명":"구동명"},axis=1)
    key1=df[["동리명"]] #df_code와 조인해서 행정동코드 뽑아내기
    key1=key1.join(df_code.df.set_index("동리명")[["구동명","행정동코드"]],on="동리명")

    df=pd.merge(df,key1,on="동리명")
    return df




##연도별 영업 상태##
#df_tot.groupby(["상세영업상태명"])["상세영업상태명"].agg("count")



path="/content/drive/My Drive/chickenPJ/data/" #경로

#####1)모든 서울시의 2010년부터 현재까지  영/폐업 
df_tot=Read_Write(path+"07_24_04_P.xlsx")
#####2)현재 영업중인 서울시 모든 치킨집 
df_pre_ch=Read_Write(path+"상가업소정보_서울_202006.xlsx")
#####3)현재 서울시 인구수 뽑아내기 최근 10년
df_pop=Read_Write(path+"행정구역_시군구_별_1세별_주민등록인구_20200817041538.xlsx")
######4)법정동 코드#####
df_code=Read_Write(path+"행정동 및 법정동 코드.xlsx")
#####5)구별 동별 좌표계
df_xy=Read_Write(path+"서울시 행정구역 시군구 정보 (좌표계_ WGS1984).csv")
df_dong_xy=Read_Write(path+"서울시 법정구역 읍면동 공간정보 (좌표계_ WGS1984).csv")

##########데이터 정제##################
df_tot=Refine.tot(df_tot.df,df_code.df)
df_pre_ch=Refine.pre_ch(df_pre_ch.df)
df_pop=Refine.pop(df_pop.df)


#########좌표계 데이터 정제############
df_code=Refine.code(df_code.df)
df_xy=Refine.xy(df_xy.df)
df_dong_xy=Refine.dong_xy(df_dong_xy.df)


########화면에 보이기############
display(df_tot,df_pre_ch,df_pop,df_code,df_xy,df_dong_xy)


```

#(2)def inc_dec_tot(df) 
창폐업율 데이터 정제

```python
#1) 연도별 시군구별 법정동별 폐업대비 창업율
def inc_dec_tot(df):
  ##df_tot를 이용해 그룹화한 후 두개의 데이터 프레임으로 조인 후 영폐업 개수 추출
  df=df.groupby(["설립연도","시군구명","동리명","상세영업상태명"])[["상세영업상태명"]].agg("count")
  df=df.rename({"상세영업상태명" : "영업상태"},axis=1)
  dfcomp=df.copy()
  df=df.reset_index()
  dfcomp=dfcomp.reset_index()
  df=df[df["상세영업상태명"]=="영업"]
  dfcomp=dfcomp[dfcomp["상세영업상태명"]=="폐업"]
  iddf=pd.merge(df,dfcomp,on=["설립연도","시군구명","동리명"],how="outer")

  #######하나로 조인한 pddf 데이터프레임으로 영 폐업 컬럼명 변경과 결측치 채우기
  iddf=iddf.rename({"상세영업상태명_x":"개업","영업상태_x":"개업개수","상세영업상태명_y":"폐업","영업상태_y":"폐업개수"},axis=1)
  iddf["개업"]=iddf["개업"].fillna("영업")
  iddf["폐업"]=iddf["폐업"].fillna("폐업")
  iddf["개업개수"]=iddf["개업개수"].fillna(0)
  iddf["폐업개수"]=iddf["폐업개수"].fillna(0)
  
  #####폐업대비 창업율 
  iddf=iddf.sort_values(["설립연도","시군구명","동리명"])
  iddf=iddf.reset_index(drop=True)
  iddf["폐업대비창업율"]=(iddf["개업개수"]-iddf["폐업개수"])/iddf["개업개수"].where(iddf["개업개수"]>0,other=1)
  return iddf




#######함수 대입#############
iddf=inc_dec_tot(df_tot)





#시군구별 연도별 개업개수 vs 폐업개수
iddx=iddf.groupby(["시군구명","설립연도"])[["개업개수","폐업개수"]].agg("sum")

iddx=iddx.reset_index()



#####################그래프 그릴것#################################

#평균개업개수 폐업개수
iddex=iddf.groupby(["시군구명"])[["개업개수","폐업개수"]].agg("mean").plot(kind="bar") #10년간
iddex.set_title("서울시 10개년 평균 개폐업 수", fontsize=20)
display(iddex)
#plt.savefig("hi.png")
#map 10 2010년 폐업 대비 창업율 지도로~ ~20년까지
```

#(4)Class Pre_Num
(현재 시군구별 치킨집 개수 지도에 나타나게 만들기)

```python
class Pre_Num:
  @staticmethod
  def Gu(df,df_xy):
    df_Gu_cnt=df.groupby("시군구명")[["동리명"]].agg("count")
    df_Gu_cnt=df_Gu_cnt.reset_index()
    df_Gu_cnt=df_Gu_cnt.rename({"동리명":"치킨집개수"},axis=1)
    pre_xy_Gu=pd.merge(df_Gu_cnt,df_xy,on="시군구명") #위도경도 합치기
    return pre_xy_Gu

  @staticmethod
  def dong(df,df_xy): #위도 경도 동별 좌표 알아내기
   df_dong=df.groupby(["시군구명","동리명","행정동코드"])["동리명"].agg("count")
   df_dong.name="치킨집개수"
   df_dong=df_dong.reset_index()
   return df_dong

dfex=Pre_Num.Gu(df_pre_ch,df_xy)

dfex


```

```python
#########지도 시각화 모듈 포리움 .맵 중심지 쓰기###########
#!pip install geopandas

import folium # 데이터
import geopandas as gpd #json 파일 읽어들여서 지도에 포리움 지도에 표시



###############map 파라매터 안에 들어갈 변수들####################
geo_data = '/content/drive/My Drive/Colab Notebooks/지도시각화(folium)/data/seoul-dong.geojson'
map_osm = folium.Map(location=[37.541, 126.986], zoom_start=11)
mapping = "행정동코드"
color="PuRd"
clr_num=9
main_name="2020년 치킨집개수"



class Map:
  def __init__(self):
    self.map_osm = folium.Map(location=[37.541, 126.986], zoom_start=11)
                    #json데이터, 데이터 프레임, 데이터
  def paint_Map(self, geo_data, df, mapping, color, clr_num, main_name):
    folium.Choropleth( 
        geo_data=geo_data, #json
        data=df, #기준 데이터 프레임
        columns=(mapping,"치킨집개수"), #x축 y축 
        key_on='feature.properties.'+mapping,  #json 키 
        fill_color=color,         #'BrBG  채우기색
        fill_opacity=0.7,       #투명도
        line_opacity=0.1,
        bins=clr_num,           #색상 수
      #nan_fill_color='white',   # nan 표시 컬러
        legend_name=main_name,  #'2020_동별치킨집개수'
        ).add_to(self.map_osm)
        

  def paint_Map_sigu(self,df): # 원그리기
    for n in df.index:
      folium.CircleMarker([df["위도"][n], df["경도"][n]],
                          radius=df['치킨집개수'][n]/3,   
                          color='#cc00cc', fill_color='#ff33ff',
                          popup = folium.Popup(
                              folium.IFrame(
                                  (
                                     "<h4>【"+df["시군구명"][n] +"】 치킨집 </h4>" + df["치킨집개수"][n].astype(np.str) +" 개"
                                  ),width=250, height=80
                                   ), max_width=200
                                           )
                          ).add_to(self.map_osm)
  def paint_Marker(self,df):
    for n in df.index:
      folium.Marker(
          [df["위도"][n], df["경도"][n]],
          popup=folium.Popup(max_width=450).add_child(
          folium.Vega(vis1, width=450, height=250))
    ).add_to(self.map_osm)



 ######### 임시로 생성한 메서드, 나중에 수정하거나 삭제해야함!!!#####################
  def paint_Map2(self,df):
    folium.Choropleth( 
        geo_data=geo_data, #json
        data=df, #기준 데이터 프레임
        columns=("시군구명","연도별평균증감"), #x축 y축 
        key_on='feature.properties.구',  #json 키 
        fill_color='PuBuGn',         #'BrBG  채우기색
        fill_opacity=0.8,       #투명도
        line_opacity=0.2,
        bins=9,           #색상 수
      #nan_fill_color='white',   # nan 표시 컬러
        legend_name="연도별평균증감",  #'2020_동별치킨집개수'
        ).add_to(self.map_osm)
                                 
  def paint_Map_sigu2(self,df):
    for n in df.index:
      folium.CircleMarker([df["위도"][n], df["경도"][n]],
                          radius=df['개업개수'][n]*12,   
                          color='#0066cc', fill_color='#3399ff',
                          ).add_to(self.map_osm)       

  def paint_Map_sigu3(self,df):
    for n in df.index:
      folium.CircleMarker([df["위도"][n], df["경도"][n]],
                          radius=df['폐업개수'][n]*12,   
                          color='#990000',
                          fill_color='#cc0000',
                          popup = folium.Popup(
                              folium.IFrame(
                                  (
                                      "<h4>【  "  +  df["시군구명"][n] +"   】치킨집</h4>"  +
                                       "10년 평균 개업개수 : "  + round(df["개업개수"][n] , 0).astype(np.str) +" 개"+ "<br>"
                                       "10년 평균 폐업개수 : "  + round(df["폐업개수"][n] , 0).astype(np.str) +" 개" + "<br>"
                                       "10년 평균 개폐업수 차이 : " +round(df["개업개수"][n] -df["폐업개수"][n] ).astype(np.str) + "개"
                                  ),width=300, height=100
                                   ), max_width=300
                                           )
                          ).add_to(self.map_osm)        
  def dp(self):
    display(self.map_osm)


#popup 옵션
#image_tag = '<img src="data:image/jpeg;base64,{}">'.format(pic)
#iframe = folium.IFrame(image_tag, width=300, height=300)
#popup = folium.Popup(folium.IFrame(image_tag, width=300, height=300), max_width=650)


############2020년 치킨집 지도########################
df_2020=df_pre_ch.groupby(["시군구명","동리명","행정동코드"])["동리명"].agg("count")
df_2020.name="치킨집개수"
df_2020=df_2020.reset_index()
a=Map()
a.paint_Map(geo_data, df_2020, mapping, color, clr_num, main_name)
a.paint_Map_sigu(dfex) #(4)에서 만든 데이터프레임 넣기


###################구별 평균 영폐업 개수##################
iddx=iddf.groupby(["시군구명"])[["개업개수","폐업개수"]].agg("mean")
iddx=iddx.reset_index()
iddx_xy=pd.merge(iddx,df_xy,on="시군구명") #위도경도 합치기
b=Map()

b.paint_Map_sigu2(iddx_xy)
b.paint_Map_sigu3(iddx_xy)


###############2020년 동별 현재 인구 상승률 평균#########################
id_pop["연도별평균증감"]=id_pop[[2011,2012,2013,2014,2015,2016,2017,2018,2019]].apply(lambda x : x.mean(),axis=1)
per_pop=pd.merge(id_pop,df_xy,on="시군구명") #df_pop_xy=pd.merge(df3,df_xy,on="시군구명",how="inner")
per_pop=per_pop[["시군구명","연도별평균증감","위도","경도"]]
c=Map()
c.paint_Map2(per_pop)



########저장################
a.map_osm.save("지도1.html")
b.map_osm.save("지도2.html")
c.map_osm.save("지도3.html")

b.map_osm
```
